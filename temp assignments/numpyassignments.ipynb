```python
arr = np.random.randint(1,21,size=(5,5))
mean = np.mean(arr)
median = np.median(arr)
std = np.std(arr)
var = np.var(arr)
print(arr)
print(f"mean: {mean}\nmedian: {median}\nstd: {std}\nvar: {var}")
```

```python
arr = np.arange(1,10).reshape((3,3))
print(f"arr: {arr}")
normalized_arr = (arr-np.mean(arr))/np.std(arr)
print(f"Normalized: {normalized_arr}")
```
